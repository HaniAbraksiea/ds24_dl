1. Hur är AI, Maskininlärning och Deep Learning relaterat?

AI är det stora området som handlar om att få datorer att tänka och agera smart. Maskininlärning är en del av AI som låter datorer lära sig från data. Deep Learning är en speciell del av maskininlärning som använder neurala nätverk och fungerar extra bra med stora datamängder.

2. Hur är Tensorflow och Keras relaterat?

TensorFlow är ett stort ramverk (bibliotek) för maskininlärning och deep learning. Keras är ett enklare gränssnitt (API) som kör ovanpå TensorFlow och gör det lättare att bygga neurala nätverk utan att skriva så mycket kod.

3. Vad är en parameter? Vad är en hyperparameter?

En parameter är något modellen lär sig själv, som vikter och biasar i ett neuralt nätverk. En hyperparameter är något vi ställer in innan träningen, som antal lager, lärandehastighet eller batch-storlek.

4. När man skall göra modellval och modellutvärdering kan man använda tränings-, validerings- och testdataset. Förklara hur de olika delarna kan användas.

Träningsdata används för att träna modellen. Valideringsdata används under träningen för att justera hyperparametrar och undvika överanpassning (overfitting). Testdata används sist, efter träningen, för att utvärdera modellens slutliga prestanda på ny, osedd data.

5. Förklara vad nedanstående kod gör:

n_cols = x_train.shape[1]

nn_model = Sequential()
nn_model.add(Dense(100, activation='relu', input_shape=(n_cols, )))
nn_model.add(Dropout(rate=0.2))
nn_model.add(Dense(50, activation='relu'))
nn_model.add(Dense(1, activation='sigmoid'))

nn_model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy' ])

early_stopping_monitor = EarlyStopping(patience=5)
nn_model.fit(
    x_train,
    y_train,
    validation_split=0.2,
    epochs=100,
    callbacks=[early_stopping_monitor])

Koden bygger ett neuralt nätverk med Keras. Det har:
en första dense-lager med 100 noder (ReLU), ett dropout-lager (för att minska överfitting), en dense-lager med 50 noder (ReLU), och en utgångslager med 1 nod (sigmoid, för binär klassificering). Modellen kompileras med Adam-optimizer och binary crossentropy som förlust. Den tränas på x_train och y_train i max 100 epoker, med tidig stoppning om valideringsprestandan inte förbättras på 5 epoker.

6. Vad är syftet med att regularisera en modell?

Syftet med att regularisera en modell är att minska risken för överanpassning (overfitting), så att modellen generaliserar bättre till ny data och inte bara lär sig träningen utantill.

7. ”Dropout” är en regulariseringsteknik, vad är det för något?

Dropout är en regulariseringsteknik där man slumpmässigt stänger av (tar bort) vissa noder i nätverket under träning, för att förhindra att modellen blir för beroende av enskilda noder och därmed minska överanpassning.

8. ”Early stopping” är en regulariseringsteknik, vad är det för något?

Early stopping innebär att man slutar träna modellen tidigt om valideringsförlusten slutar förbättras, för att undvika att modellen överanpassar till träningsdatan.

9. Din kollega frågar dig vilken typ av neuralt nätverk som är populärt för bildanalys, vad svarar du?

Jag skulle svara att konvolutionella neurala nätverk (CNN) är den vanligaste typen för bildanalys, eftersom de är bra på att upptäcka mönster i bilder.

10. Förklara översiktligt hur ett ”Convolutional Neural Network” fungerar.

Ett Convolutional Neural Network (CNN) fungerar genom att först använda konvolutionslager som söker efter mönster (som kanter eller former) i bilden. Sedan används pooling-lager för att minska storleken på datan. Till slut används dense-lager för att göra en klassificering baserat på de upptäckta mönstren.

12. Deep Learning modeller kan ta lång tid att träna, då kan GPU via t.ex. Google Colab skynda på träningen avsevärt. Skriv mycket kortfattat vad CPU och GPU är.

CPU (Central Processing Unit) är datorns allmänna hjärna, bra på många små uppgifter.
GPU (Graphics Processing Unit) är specialiserad på att göra många beräkningar parallellt, vilket gör den snabbare för träning av deep learning-modeller.
